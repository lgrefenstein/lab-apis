```python
## Collect Pokémon data
pokemon_list = []

for i in range(1, 152): 
    url = f"https://pokeapi.co/api/v2/pokemon/{i}"
    
    response = requests.get(url)
    data = response.json()
    
    pokemon_list.append(data)
```

```python
len(pokemon_list)
```

```python
## Extract relevant features
clean_data = []

for pokemon in pokemon_list:
    
    name = pokemon["name"]
    height = pokemon["height"]
    weight = pokemon["weight"]
    base_experience = pokemon["base_experience"]
    
    first_type = pokemon["types"][0]["type"]["name"]
    
    clean_data.append({
        "name": name,
        "height": height,
        "weight": weight,
        "base_experience": base_experience,
        "type": first_type
    })
```

```python
clean_data[:3]
```

```python
## Create the DataFrame
pokemon_df = pd.DataFrame(clean_data)

pokemon_df.head()
```

```python
pokemon_df.shape
```

```python
pokemon_df.dtypes
```

```python
## Create a weight column in kg
pokemon_df["weight_kg"] = pokemon_df["weight"] / 10
```

```python
pokemon_df.head()
```

Analysis

```python
## Heaviest Pokémon
pokemon_df.loc[pokemon_df["weight_kg"].idxmax()]
```

```python
## Average weight
pokemon_df.groupby("type")["weight_kg"].mean().sort_values(ascending=False)
```

```python
## Distribution by types
pokemon_df["type"].value_counts()
```

```python
## Compared average weight across types
pokemon_df.groupby("type")["weight_kg"].mean().sort_values(ascending=False)
```

```python
import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/simple/price"

params = {
    "ids": "bitcoin,ethereum,solana,dogecoin",
    "vs_currencies": "usd"
}

response = requests.get(url, params=params)
data = response.json()

data
```

```python
## 2. Convert API response into DataFrame
crypto_df = pd.DataFrame(data).T
crypto_df
```

```python
crypto_df.shape
```

```python
crypto_df.columns
```

```python
## Rename columns
crypto_df = crypto_df.rename(columns={"usd": "price"})
crypto_df
```

```python
## Add a price rank
crypto_df["price_rank"] = crypto_df["price"].rank(ascending=False).astype(int)
crypto_df
```

```python
## Add a price category
def price_category(price):
    if price > 1000:
        return "high"
    elif price >= 100:
        return "medium"
    else:
        return "low"

crypto_df["price_category"] = crypto_df["price"].apply(price_category)
crypto_df
```

```python
crypto_df["price_category"].value_counts()
```

Analysis

```python
## Most expensive coin
crypto_df.loc[crypto_df["price"].idxmax()]
```

```python
## Which category has more coins?
crypto_df["price_category"].value_counts()
```

```python
## Price difference vs Bitcoin
bitcoin_price = crypto_df.loc["bitcoin", "price"]

crypto_df["price_diff_vs_btc"] = crypto_df["price"] - bitcoin_price

crypto_df
```
